```python
import os
from mpasanalysis import *
%matplotlib inline
```

```python
casename = '20190807.DECKv1b_P1_SSP5-8.5.ne30_oEC.cori-knl'
dataroot = os.environ['HOME']+'/data/mpas/cmip6'
# filename = 'mpaso.hist.am.sst.highFrequencyOutput.2015-2101.nc'
filename = 'mpaso.hist.am.ke.highFrequencyOutput.2015-2101.nc'
inputfile = dataroot + '/' + casename + '/' + filename
meshroot = os.environ['HOME']+'/data/mpas/grids'
meshname = 'oEC60to30v3_60layer.170905.meshonly.nc'
meshfile = meshroot + '/' + meshname
figroot = os.environ['HOME']+'/work/cmip6/TIW/figures'
```

```python
mpasdata = MPASOData(filepath=inputfile, filepath_mesh=meshfile)
```

```python
ncdata = mpasdata.load()
xtime = chartostring(ncdata.variables['xtime'][:])
time = [x.strip()[:10] for x in xtime]
yyyy = np.array([float(tmp[0:4]) for tmp in time])
ntime = len(time)
print(ntime)
yrs = 2015
yre = 2020
```

```python
region1 = 'TropicalPacific'
region2 = 'TropicalAtlantic'
save_fig = True

# levels1 = np.linspace(16,31,76)
# levels2 = np.linspace(20,32,61)
# varname = 'temperatureAtSurface'
# var = 'SST'

levels1 = np.linspace(0,1.5,76)
levels2 = np.linspace(0,0.6,61)
varname = 'kineticEnergyAtSurface'
name = 'KE'
units = 'm^2/s^2'

for i in np.arange(ntime):
# for i in [6000, 6100, 6200]:
# for i in [6, 100, 150]:
    # filter year
    if yyyy[i] < yrs or yyyy[i] > yre :
        continue
        
    # load data
    print(time[i])
    sst = mpasdata.get_map(varname=varname, name=name, units=units, tidx=i)
    
    # plot figure
    fig = plt.figure(figsize=[12, 3.5])
    m, img = sst.plot(region=region1, levels=levels1, title=time[i], projection='cyl')
    if save_fig:
        figname = figroot+'/TIW_'+name+'_'+region1+'_'+time[i]+'.png'
        plt.tight_layout()
        fig.savefig(figname, dpi = 300)
        plt.close(fig)
    
    fig = plt.figure(figsize=[5, 3.5])
    m, img = sst.plot(region=region2, levels=levels2, title=time[i], projection='cyl')
    if save_fig:
        figname = figroot+'/TIW_'+name+'_'+region2+'_'+time[i]+'.png'
        plt.tight_layout()
        fig.savefig(figname, dpi = 300)
        plt.close(fig)
```
