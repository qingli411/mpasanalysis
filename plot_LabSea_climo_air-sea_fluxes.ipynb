```python
from mpasanalysis import *
import e3sm_res_cmp
%matplotlib inline
```

```python
# get paths of restart files, monthly mean output files, processed climatology files and output figures
runname = 'low-res-cmp'
data_root = e3sm_res_cmp.load_paths_ocn(climo_ys=21, climo_ye=50, ts_ys=1, ts_ye=50, runname=runname)
rst_root = data_root['rst_root']
climo_root = data_root['climo_root']
fig_root = data_root['fig_root']
rst_file = rst_root+'/mpaso.rst.0051-01-01_00000.nc'
data_root_ice = e3sm_res_cmp.load_paths_ice(climo_ys=21, climo_ye=50, ts_ys=1, ts_ye=50, runname=runname)
climo_root_ice = data_root_ice['climo_root']
```

```python
# load dataset
mpasmesh = MPASMesh(filepath=rst_file)
```

## heat flux

```python
levels = np.linspace(-400, 400, 81)
levels_ifrac = np.array([0.15, 0.85])
for i in np.arange(12):
    climo_file = climo_root+'/heatFlux/mpaso_{:02d}_climo.nc'.format(i+1)
    mpasodata = MPASOData(filepath=climo_file, filepath_mesh=rst_file)
    f_climo = mpasodata.load()
    shf = f_climo.variables['timeMonthly_avg_sensibleHeatFlux'][0,:]
    lhf = f_climo.variables['timeMonthly_avg_latentHeatFlux'][0,:]
    swf = f_climo.variables['timeMonthly_avg_shortWaveHeatFlux'][0,:]
    lwfu = f_climo.variables['timeMonthly_avg_longWaveHeatFluxUp'][0,:]
    lwfd = f_climo.variables['timeMonthly_avg_longWaveHeatFluxDown'][0,:]
    flux_tot = shf+lhf+swf+lwfu+lwfd
    mpaso_flux = MPASOMap(data=flux_tot, mesh=mpasmesh,
                          name='Total Heat Flux', units=r'W m^{-2}')
    climo_file_ifrac = climo_root_ice+'/iceFraction/mpascice_{:02d}_climo.nc'.format(i+1)
    mpascicedata = MPASCICEData(filepath=climo_file_ifrac, filepath_mesh=rst_file)
    mpascice_ifrac = mpascicedata.get_map(varname='timeMonthly_avg_iceAreaCell',
                                           name='Ice fraction', units='none')
    
    # Temperature map at different depths
    fig = plt.figure(figsize=[6,6])
    m, f = mpaso_flux.plot(region='LabSea', ptype='contourf', levels=levels, cmap='RdBu_r')
    mpascice_ifrac.overlay(m, levels=levels_ifrac, cmap='bone_r')
    plt.show()
    figname = fig_root+'/LabSea_climo_{:02d}_Map_total_heat_flux.png'
    fig.savefig(figname, dpi = 300)
    plt.close(fig)
        
```

## Salinity

```python
levels = np.linspace(-2e-4, 2e-4, 41)
levels_ifrac = np.array([0.15, 0.85])
for i in np.arange(12):
    climo_file = climo_root+'/freshWaterFlux/mpaso_{:02d}_climo.nc'.format(i+1)
    mpasodata = MPASOData(filepath=climo_file, filepath_mesh=rst_file)
    f_climo = mpasodata.load()
    evap = f_climo.variables['timeMonthly_avg_evaporationFlux'][0,:]
    rain = f_climo.variables['timeMonthly_avg_rainFlux'][0,:]
    snow = f_climo.variables['timeMonthly_avg_snowFlux'][0,:]
    si_s = f_climo.variables['timeMonthly_avg_seaIceSalinityFlux'][0,:]
    si_f = f_climo.variables['timeMonthly_avg_seaIceFreshWaterFlux'][0,:]
    ro_r = f_climo.variables['timeMonthly_avg_riverRunoffFlux'][0,:]
    ro_i = f_climo.variables['timeMonthly_avg_iceRunoffFlux'][0,:]
    flux_tot = evap+rain+snow+si_s+si_f+ro_r+ro_i
    mpaso_flux = MPASOMap(data=flux_tot, mesh=mpasmesh,
                         name='Total Fresh Water Flux', units=r'kg m^{-2} s^{-1}')
    
    climo_file_ifrac = climo_root_ice+'/iceFraction/mpascice_{:02d}_climo.nc'.format(i+1)
    mpascicedata = MPASCICEData(filepath=climo_file_ifrac, filepath_mesh=rst_file)
    mpascice_ifrac = mpascicedata.get_map(varname='timeMonthly_avg_iceAreaCell',
                                           name='Ice fraction', units='none')
    
    fig = plt.figure(figsize=[6,6])
    m, f = mpaso_flux.plot(region='LabSea', ptype='contourf', levels=levels, cmap='RdBu_r')
    mpascice_ifrac.overlay(m, levels=levels_ifrac, cmap='bone_r')
    plt.show()
    figname = fig_root+'/LabSea_climo_{:02d}_Map_fresh_water_flux.png'
    fig.savefig(figname, dpi = 300)
    plt.close(fig)
        
```
