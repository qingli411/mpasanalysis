```python
mpas_data = MPASOData(filepath=data_file, filepath_mesh=mesh_file)
```

### plot mesh

```python
fig = plt.figure(figsize=[10,10])
mpas_data.mesh.plot_celledges_xy()
xshift = -1000
yshift = 40000
plt.xlim([0+xshift, 10000+xshift])
plt.ylim([0+yshift, 10000+yshift])
```

```python
def plot_hvlines(xfrac, yfrac, **kwargs):
    ax = plt.gca()
    xmin, xmax = ax.get_xlim()
    ymin, ymax = ax.get_ylim()
    xf = xmin+xfrac*(xmax-xmin)
    yf = ymin+yfrac*(ymax-ymin)
    plt.axvline(xf, **kwargs)
    plt.axhline(yf, **kwargs)
```

### plot temperature

```python
temp = mpas_data.get_domain(varname='temperature', position='cell', name='temperature', units='degC', tidx=20)
```

```python
xfrac = 0.5
yfrac = 0.5
```

```python
fig = plt.figure(figsize=[12,10])
levels=np.linspace(21.4,22,31)
temp.plot_xy(ptype='contourf', levels=levels)
plot_hvlines(xfrac, yfrac, color='k')
# temp.plot_xy(ptype='contourf')
# plt.xlim([0, 10000])
# plt.ylim([20000, 30000])
```

```python
fig = plt.figure(figsize=[12,6])
levels=np.linspace(20,22,41)
temp.plot_yz(xfrac=xfrac, ptype='contourf', levels=levels)
plt.ylim([-100,0])
```

```python
fig = plt.figure(figsize=[12,6])
temp.plot_xz(yfrac=yfrac, ptype='contourf', levels=levels)
plt.ylim([-100,0])
```
